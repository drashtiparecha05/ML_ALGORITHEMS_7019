# All type linkage cluster

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.metrics import adjusted_rand_score

data = load_iris()
X, y = data.data, data.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Single Linkage
Z_single = linkage(X_scaled, method='single')
labels_single = fcluster(Z_single, t=3, criterion='maxclust')
print("Single Linkage ARI:", adjusted_rand_score(y, labels_single))

plt.figure(figsize=(12, 4))
plt.title('Dendrogram - Single Linkage')
dendrogram(Z_single, truncate_mode='level', p=5)
plt.show()
```

```python
# Average Linkage
Z_average = linkage(X_scaled, method='average')
labels_average = fcluster(Z_average, t=3, criterion='maxclust')
print("Average Linkage ARI:", adjusted_rand_score(y, labels_average))

plt.figure(figsize=(12, 4))
plt.title('Dendrogram - Average Linkage')
dendrogram(Z_average, truncate_mode='level', p=5)
plt.show()
```

```python
# Complete Linkage
Z_complete = linkage(X_scaled, method='complete')
labels_complete = fcluster(Z_complete, t=3, criterion='maxclust')
print("Complete Linkage ARI:", adjusted_rand_score(y, labels_complete))

plt.figure(figsize=(12, 4))
plt.title('Dendrogram - Complete Linkage')
dendrogram(Z_complete, truncate_mode='level', p=5)
plt.show()
```

```python
# Comparison Plot
from sklearn.decomposition import PCA
X_pca = PCA(n_components=2).fit_transform(X_scaled)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, labels, title in zip(axes,
    [labels_single, labels_average, labels_complete],
    ['Single', 'Average', 'Complete']):
    ax.scatter(X_pca[:,0], X_pca[:,1], c=labels, cmap='Set1', alpha=0.7)
    ax.set_title(f'{title} Linkage')
plt.tight_layout()
plt.show()
```
