# Association Rules market basket

```python
# Install mlxtend if not available
import subprocess
subprocess.run(['pip', 'install', 'mlxtend', '-q'])
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Sample market basket dataset
dataset = [
    ['Milk', 'Bread', 'Butter'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread'],
    ['Milk', 'Bread', 'Butter', 'Eggs'],
    ['Bread', 'Eggs'],
    ['Milk', 'Eggs'],
    ['Milk', 'Bread', 'Eggs'],
    ['Butter', 'Eggs'],
    ['Milk', 'Butter'],
    ['Bread', 'Butter', 'Eggs'],
    ['Milk', 'Bread', 'Butter', 'Eggs'],
    ['Milk', 'Bread'],
    ['Butter', 'Eggs'],
    ['Milk', 'Bread', 'Butter'],
    ['Bread', 'Eggs'],
]

te = TransactionEncoder()
te_array = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_array, columns=te.columns_)
print("Transaction Matrix:")
print(df.head())
```

```python
# Frequent Itemsets using Apriori
frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
print("\nFrequent Itemsets:")
print(frequent_itemsets.sort_values('support', ascending=False))
```

```python
# Association Rules
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
rules = rules.sort_values('confidence', ascending=False)
print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10))
```

```python
import matplotlib.pyplot as plt

# Support vs Confidence scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(rules['support'], rules['confidence'], alpha=0.6, c=rules['lift'], cmap='viridis')
plt.colorbar(label='Lift')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Association Rules - Support vs Confidence (color = Lift)')
plt.show()
```
