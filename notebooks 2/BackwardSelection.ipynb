# BackwardSelection

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np

df = pd.read_csv('cputime.csv')
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

selected = list(X.columns)

print("Backward Feature Selection:")
while len(selected) > 1:
    scores = {}
    for feature in selected:
        current = [f for f in selected if f != feature]
        model = LinearRegression()
        score = cross_val_score(model, X[current], y, cv=5, scoring='r2').mean()
        scores[feature] = score
    
    worst_feature = max(scores, key=scores.get)
    print(f"  Removed: {worst_feature} | R2 without it: {scores[worst_feature]:.4f}")
    selected.remove(worst_feature)

print("\nFinal selected features:", selected)
```
