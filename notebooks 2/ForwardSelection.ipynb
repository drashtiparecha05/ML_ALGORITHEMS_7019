# ForwardSelection

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np

df = pd.read_csv('cputime.csv')
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

selected = []
remaining = list(X.columns)
best_score = -np.inf

print("Forward Feature Selection:")
while remaining:
    scores = {}
    for feature in remaining:
        current = selected + [feature]
        model = LinearRegression()
        score = cross_val_score(model, X[current], y, cv=5, scoring='r2').mean()
        scores[feature] = score
    
    best_feature = max(scores, key=scores.get)
    if scores[best_feature] > best_score:
        best_score = scores[best_feature]
        selected.append(best_feature)
        remaining.remove(best_feature)
        print(f"  Added: {best_feature} | R2: {best_score:.4f}")
    else:
        break

print("\nFinal selected features:", selected)
```
