# kMeans KMedoid Iris

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score
import matplotlib.pyplot as plt

# Try to use KMedoids if mlxtend is available, else fallback
try:
    from mlxtend.cluster import Kmeans as MedoidKMeans
    print("mlxtend available")
except:
    print("mlxtend not available, using KMeans only")

data = load_iris()
X, y = data.data, data.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)
print("KMeans ARI:", adjusted_rand_score(y, labels))

plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels, cmap='Set1', alpha=0.7)
plt.title('KMeans - Iris Dataset')
plt.xlabel('Sepal Length (scaled)')
plt.ylabel('Sepal Width (scaled)')
plt.show()
```
